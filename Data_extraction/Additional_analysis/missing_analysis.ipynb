```python
import numpy as np
import pandas as pd
import os 
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

pd.set_option('display.max_rows', 200)
```

```python
df = pd.read_parquet("../Data/train_data_split.parquet")
print(df.shape)
df.head(5)
```

```python
for column in df.columns:
    df[column] = df[column].replace(-9999, np.nan)
```

```python
df.head(5)
```

## Analyze missing values

```python
print("Does target variable have missing values: ", df.target.isna().any())
```

```python
for column in df.columns:
    missing_percentage = round(np.mean(df[column].isna()),2)*100
    if missing_percentage != 0:
        print("Variable: {} has {}% missing values".format(column, round(missing_percentage,2)))
```

```python
for column in df.columns:
    missing_percentage = np.mean(df[column].isna())*100
    if missing_percentage > 95:
        print("Variable: {} has {}% missing values".format(column, round(missing_percentage,2)))
```

```python
for column in df.columns:
    missing_percentage = round(np.mean(df[column].isna()),2)*100
    if missing_percentage != 0:
        column_missings = df.loc[:,[column, "customer_obs"]].groupby(["customer_obs"]).agg(lambda x: sum(x.isna()))
        column_missings["non_missing"] = df.loc[:,[column, "customer_obs"]].groupby(["customer_obs"]).count()
        column_missings["mis_%"] = round(
            column_missings[column] / (column_missings[column] + column_missings["non_missing"])*100, 2)
        print(column_missings)
```

## Analysis of target variable distributions depending on occurrence of missing value

```python
y = df.loc[:,["customer_ID", "target"]].groupby("customer_ID").mean().reset_index()
y["target"] = y["target"].astype(int)
```

#### Check the distribution of Y for customers that have at least 1 missing in a particular variable

```python
for column in df.columns:
    missing_percentage = round(np.mean(df[column].isna()),4)*100
    if missing_percentage != 0:
        customer_miss = pd.DataFrame(
            df.loc[:,["customer_ID", column]].groupby("customer_ID")[column].apply(
                                                                        lambda x: x.isnull().any())).reset_index()
        
        customer_full = customer_miss[~customer_miss[column]].reset_index(drop=True)
        customer_missing = customer_miss[customer_miss[column]].reset_index(drop=True)
        customer_missing = customer_missing.merge(y, on="customer_ID", how="left")
        customer_full = customer_full.merge(y, on="customer_ID", how="left")
        
        print("For: ", column)
        print("Variable: {} has {}% missing values".format(column, round(missing_percentage,2)))
        print("Y_full: ", round(customer_full.target.mean(),2))
        print("Y_miss: ", round(customer_missing.target.mean(),2))
        
```

#### Check the distribution of Y for customers that have all missing observations in a particular variable

```python
for column in df.columns:
    missing_percentage = round(np.mean(df[column].isna()),4)*100
    if missing_percentage != 0:
        customer_miss = pd.DataFrame(
            df.loc[:,["customer_ID", column]].groupby("customer_ID")[column].apply(
                                                                        lambda x: x.isnull().all())).reset_index()
        
        customer_full = customer_miss[~customer_miss[column]].reset_index(drop=True)
        customer_missing = customer_miss[customer_miss[column]].reset_index(drop=True)
        customer_missing = customer_missing.merge(y, on="customer_ID", how="left")
        customer_full = customer_full.merge(y, on="customer_ID", how="left")
        
        print("For: ", column)
        print("Y_full: ", round(customer_full.target.mean(),2))
        print("Y_miss: ", round(customer_missing.target.mean(),2))
```
